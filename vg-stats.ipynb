## Read CSV File

```python
df = pd.read_csv('data/vgsales.csv')
```

```python
df
```

## Which company is the most common video game publisher ?

```python
sorted_publisher=df.pivot_table(columns=['Publisher'], aggfunc='size')
sorted_publisher=sorted_publisher.sort_values( ascending = False )
sorted_publisher.head(1)
```

## What’s the most common platform ?

```python
plat_form = df ['Platform'].value_counts()
plat_form.head(1)
```

## What about the most common genre ?

```python
gen = df ['Genre'].value_counts()
gen.head(1)
```

## What are the top 20 highest grossing games ?

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms ?

```python
avgs= df [ ['Platform','Global_Sales'] ].groupby('Platform').mean().sort_values('Global_Sales' , ascending = False ).head(10)
avgs
```
